```python
# Converting object data type to numeric
def convert_to_numeric(df):
    num_col = df.shape[1]
    # Start from index 3
    for index in range(3, num_col):
        col_name = df.iloc[:, index].name
        df[col_name] = pd.to_numeric(df[col_name], errors="coerce")
    return df

df = convert_to_numeric(df)
df.dtypes
```

```python
df.isnull().sum()
```

```python
df_num = df.select_dtypes(exclude="object")
```

```python
df['Temp']= df.groupby("STATE")['Temp'].fillna(df['Temp'].mean())
df['D.O. (mg/l)']= df.groupby("STATE")['D.O. (mg/l)'].fillna(df['D.O. (mg/l)'].mean())
df['PH']= df.groupby("STATE")['PH'].fillna(df['PH'].mean())
df['CONDUCTIVITY (µmhos/cm)']= df.groupby("STATE")['CONDUCTIVITY (µmhos/cm)'].fillna(df['CONDUCTIVITY (µmhos/cm)'].mean())
df['B.O.D. (mg/l)']= df.groupby("STATE")['B.O.D. (mg/l)'].fillna(df['B.O.D. (mg/l)'].mean())
df['NITRATENAN N+ NITRITENANN (mg/l)']= df.groupby("STATE")['NITRATENAN N+ NITRITENANN (mg/l)'].fillna(df['NITRATENAN N+ NITRITENANN (mg/l)'].mean())
df['FECAL COLIFORM (MPN/100ml)']= df.groupby("STATE")['FECAL COLIFORM (MPN/100ml)'].fillna(df['FECAL COLIFORM (MPN/100ml)'].mean())
df['TOTAL COLIFORM (MPN/100ml)Mean']= df.groupby("STATE")['TOTAL COLIFORM (MPN/100ml)Mean'].fillna(df['TOTAL COLIFORM (MPN/100ml)Mean'].mean())
```
